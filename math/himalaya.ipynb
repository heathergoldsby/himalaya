```python
import research as r
import scipy
import pandas


def quick_benchmark(path, filename):
    W = r.load_files(r.find_files(path, filename))

    figure()
#    for i,g in W.groupby('trial'):
#        plot(g['update'], g['max_fitness'])
        
    r.quick_ciplot('update', 'mean_fitness', W)
    xlabel("update")
    ylabel("fitness")
    
def quick_look(alps,qhfc,himalaya):
    P = ["f101", "griewangk", "n32k8", "rana", "rosen"]
    
    # alps
    for p in P:
        quick_benchmark(alps, p+".*/meta_population_fitness.dat")
        title("alps: " + p)
        
    # qhfc
    for p in P:
        quick_benchmark(qhfc, p+".*/meta_population_fitness.dat")
        title("qhfc: " + p)
        
    # himalaya
    for p in P:
        quick_benchmark(himalaya, p+".*/fitness.dat")
        title("himalaya: " + p)
```

```python
v = "/Users/dk/research/src/himalaya/var/"
quick_look(v+"001-alps", v+"002-qhfc", v+"003-himalaya")
```
