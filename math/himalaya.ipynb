```python
import research as r
import scipy
import pandas


def quick_benchmark(path, filename):
    W = r.load_files(r.find_files(path, filename))
    W = W[W['evaluation']<1000000]
        
    figure()
    r.quick_ciplot('evaluation', 'min_fitness', W)
    r.quick_ciplot('evaluation', 'mean_fitness', W)
    r.quick_ciplot('evaluation', 'max_fitness', W)
    xlabel("evaluation")
    ylabel("fitness")
    
def quick_look(alps,qhfc,himalaya):
    P = ["f101", "griewangk", "n32k8", "rana", "rosen"]
    
    # alps
    for p in P:
        quick_benchmark(alps, p+".*/fitness_evaluations.dat")
        title("alps: " + p)
        quick_benchmark(qhfc, p+".*/fitness_evaluations.dat")
        title("qhfc: " + p)
        quick_benchmark(himalaya, p+".*/fitness_evaluations.dat")
        title("himalaya: " + p)
```
